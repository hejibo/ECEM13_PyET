# High Level ioHub API Review

### Starting the ioHub Server Process

There are two ways to create a psychopy experiment and have it start the iohub process for use during the experiment runtime. Both menthods ultimately give you access to an instance of the [ioHubConnection Class)[http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/getting_connected.html#the-iohubconnection-class] for communication and control of the iohub framework:

1. Use the [psychopy.iohub.launchHubServer()](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/launchHubServer.html#the-launchhubserver-function] function.)
2. Create a subclass of the [psychopy.iohub.client.ioHubExperimentRuntime](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/ioHubExperimentRuntime.html#the-iohubexperimentruntime-class), implementing the classes run() method for your experiment's starting python code.

Each approach has +'s and -'s:

- Using launchHubServer is quicker when creating an experiment that uses simple devices like the mouse and keyboard. 
- When using more advanced devices like an eye tracker, using the ioHubExperimentRuntime approach quickly becomes easier to manage, as long as you understand the simple file structure expected by the ioHubExperimentRuntime.
** 

#### Using [launchHubServer()](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/launchHubServer.html#the-launchhubserver-function)



```python
def startStopHub(**kwargs):
    """
    Starts the ioHub Process using launchHubServer. When started with no key word arguements, 
    the ioHub Process is started with Display, Keyboard, Mouse, and Experiment devices created 
    using their default configurations. 
    The ioDataStore is disabled by default, so no events are saved by the ioHub process. 

    Optional key word arguements:

    session_code: A string providing the code to be assigned to the current run / session of the experiment. 
                  If provided, the ioDataStore is also initialized for event storage.

    experiment_code: A string providing the code to be given to the experiment data stored in the ioDataStore.
                If experiment_code is provided and session_code is not, then session_code is auto generated
                using::
                
                import psychopy.data
                session_code="S_{0}".format(psychopy.data.getDateStr()) 

    psychopy_monitor_name: Specifies the psychopy monitor configuration file to read the the display physical size
                and the eye to display distance to use in ioHub event calculations.

    datastore_name: The name to give the HDF5 file for the experiment if, based on other keyword arguements provided,
                the ioDataStore will be ective during the experiment. A default name is used if needed.

    [iohub_device_class_name]: Any key word arguenet that equals the name of any ioHub Device class, is assumed
                to indicate that you are providing a preferences dictionary containing any non default preference
                values to be used during the experiment.
    """
    io=psychopy.iohub.launchHubServer(**kwargs)

    keyboard=io.devices.keyboard

    display("Press any Key to Stop the ioHub Service:")
    
    keyboard.clearEvents()
    
    while True:
        kb_events=keyboard.getEvents() 
        if kb_events:
            display("Keyboard Event(s) Detected: ")
            for e in kb_events:
                display(e)
            break
        display('.')
        sys.stdout.flush()
        io.wait(0.5)

    io.quit() 

if __name__ == '__main__':
    startStopHub()
    #startStopHub(experiment_code='silly_exp')
    #startStopHub(experiment_code='silly_exp',session_code='s_001_m_42')
    #exp_kwargs={'experiment_code':'silly_exp','session_code'='s_001_m_42','xinput.Gamepad'={}}
    #startStopHub(**exp_kwargs)
    
```

### Extending the [ioHubExperimentRuntime class](http://www.isolver-solutions.com/iohubdocs/iohub/api_and_manual/iohub_process/ioHubExperimentRuntime.html#the-iohubexperimentruntime-class)

When the ioHubExperimentRuntime class is used to enable the ioHub Process for your experiment, the following *experiment file structure* must be used:

- The Python source file containing the ioHubExperimentRuntime subclass for your experiment must be in a directory dedicated to that experiment (lets refer to it as the exp_dir).
- The exp_dir folder must also contain two configuration files:

    - An experiment_config.yaml file specifying experiment and session level settings.
    - An iohub_config.yaml file that mainly provides the configuration information for each ioHub Device that should be enabled for the experiment.

```python
class ExperimentRuntime(psychopy.iohub.ioHubExperimentRuntime):
    """
    Create an experiment using psychopy and the ioHub framework by extending 
    the ioHubExperimentRuntime class. At minimum all that is needed is to override
    the run() method with the main experiment script logic.
    """
    def run(self,*args):
        """
        The run method contains your experiment logic. 
        It is equal to what would be in your main psychopy experiment
        script .py file in a standard psychopy experiment setup. 
        """

        # Clear all events from the global and device level event buffers.
        self.hub.clearEvents('all')
        display("End Demo By Pressing the Space Key:")
        run_demo=True
        while run_demo:
            for k in self.devices.keyboard.getEvents(EventConstants.KEYBOARD_PRESS):
                if k.key == u' ':
                    run_demo=False
                    break
                display(k)
            self.hub.wait(0.25)
        ### End of experiment logic

###############################################################################

if __name__ == "__main__":
    def main(configurationDirectory):
        """
        Creates an instance of the ExperimentRuntime class, 
        and launches the experiment logic in run().
        """
        runtime=ExperimentRuntime(configurationDirectory, "experiment_config.yaml")    
        runtime.start(sys.argv)
    # run the main function, which starts the experiment runtime
    main(module_directory(main))
```
