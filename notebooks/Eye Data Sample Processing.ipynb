#1. Pixel -> Visual Degrees Calculation


```python
def testPix2Angle(trial_samples):    
    vac=VisualAngleCalc(display_size_mm=(500.0,300.0),display_res_pix=(1280.0,1024.0),eye_distance_mm=550.0)
  
    left_x_deg,left_y_deg=vac.pix2deg(left_x,left_y)
    right_x_deg,right_y_deg=vac.pix2deg(right_x,right_y)

    leftminpv=min(left_x.min(),left_y.min())
    leftmaxpv=max(left_x.max(),left_y.max())
    rightminpv=min(right_x.min(),right_y.min())
    rightmaxpv=max(right_x.max(),right_y.max())
    leftminav=min(left_x_deg.min(),left_y_deg.min())
    leftmaxav=max(left_x_deg.max(),left_y_deg.max())
    rightminav=min(right_x_deg.min(),right_y_deg.min())
    rightmaxav=max(right_x_deg.max(),right_y_deg.max())

    copy_left_x=left_x.copy()
    copy_left_y=left_y.copy()
    copy_right_x=right_x.copy()
    copy_right_y=right_y.copy()
    copy_left_x_deg=left_x_deg.copy()
    copy_left_y_deg=left_y_deg.copy()
    copy_right_x_deg=right_x_deg.copy()
    copy_right_y_deg=right_y_deg.copy()
    
    copy_left_x[missing_left_indexes]=np.NaN
    copy_left_y[missing_left_indexes]=np.NaN
    copy_right_x[missing_right_indexes]=np.NaN
    copy_right_y[missing_right_indexes]=np.NaN
    copy_left_x_deg[missing_left_indexes]=np.NaN
    copy_left_y_deg[missing_left_indexes]=np.NaN
    copy_right_x_deg[missing_right_indexes]=np.NaN
    copy_right_y_deg[missing_right_indexes]=np.NaN    
    
    # scale angle data so it fills pixel y axis space in plot
    #
    left_angle_scalar=(leftmaxpv-leftminpv)/(leftmaxav-leftminav)
    right_angle_scalar=(rightmaxpv-rightminpv)/(rightmaxav-rightminav)
    
    plotResults(trial_id,time,(copy_left_x,copy_left_y,left_pupil,copy_right_x,copy_right_y,right_pupil),
                    (copy_left_x_deg*left_angle_scalar,copy_left_y_deg*left_angle_scalar,copy_right_x_deg*right_angle_scalar,copy_right_y_deg*right_angle_scalar),
                    fig_title='Pixel and Scaled (x%.2f) Visual Degree Positions'%(left_angle_scalar,),ylabel='Position',ds1_label_pref='Pixel',ds2_label_pref='Visual Degrees')
    
    return left_x_deg,left_y_deg,right_x_deg,right_y_deg

left_x_deg,left_y_deg,right_x_deg,right_y_deg=testPix2Angle(sample_data[trial_id])
```

#2. Calculating Eye Sample Noise

```python
def calcSampleStreamNoise():
    left_velocity_x=calculateVelocity(time,left_x_deg)
    left_velocity_y=calculateVelocity(time,left_y_deg)

#    left_x_deg_copy=left_x_deg.copy()
#    left_y_deg_copy=left_y_deg.copy()
#    left_velocity_x_copy=left_velocity_x.copy()
#    left_velocity_y_copy=left_velocity_y.copy()

    left_velocity=np.sqrt(left_velocity_x*left_velocity_x+left_velocity_y*left_velocity_y)

#    left_x_deg_copy=left_x_deg_copy[~missing_left_indexes]
#    left_y_deg_copy=left_y_deg_copy[~missing_left_indexes]
#    left_velocity_x_copy=left_velocity_x_copy[~missing_left_indexes[1:-1]]
#    left_velocity_y_copy=left_velocity_y_copy[~missing_left_indexes[1:-1]]
    left_velocity_trimmed=np.absolute(left_velocity[~missing_left_indexes[1:-1]])


    import matplotlib.pyplot as plt
    import matplotlib.mlab as mlab
    
    # filter 
    max_velocity_thresh=scipy.stats.scoreatpercentile(left_velocity_trimmed,50)
#    print 'max_velocity_thresh: ',max_velocity_thresh,left_velocity_trimmed.max(),left_velocity_trimmed.std(),np.median(left_velocity_trimmed)
    keep_left_all=left_velocity<=max_velocity_thresh
#    keep_left=left_velocity_trimmed<=max_velocity_thresh
  
    # calculate STDEV and RMS based noise Noise
    notmasked_contiguous=np.ma.extras.notmasked_contiguous
    marray=np.ma.array
    left_x_invalid_data_periods=notmasked_contiguous(marray(left_x_deg[1:-1], mask=keep_left_all))
#    print '-----'
    for data_slice in left_x_invalid_data_periods:
        if np.any(left_pupil[data_slice.start:data_slice.stop][left_pupil[data_slice.start:data_slice.stop]==0]):
            pass
        elif data_slice.stop-data_slice.start<=2:
            print data_slice,data_slice.stop-data_slice.start,keep_left_all[data_slice.start:data_slice.stop]
            keep_left_all[data_slice.start:data_slice.stop]=True
#    print '--done--'    
    left_x_valid_data_periods=notmasked_contiguous(marray(left_x_deg[1:-1], mask=~keep_left_all))

    stdev_x=0
    stdev_y=0
    rms_x=0
    rms_y=0
    for data_slice in left_x_valid_data_periods:
        print data_slice,data_slice.stop-data_slice.start
        region_length=data_slice.stop-data_slice.start
        
        if np.any(left_pupil[data_slice.start:data_slice.stop][left_pupil[data_slice.start:data_slice.stop]==0]):
            pass
        elif region_length==1:
            #print keep_left_all[data_slice.start]
            keep_left_all[data_slice.start]=False
        else:
            stdev_x+=left_x_deg[data_slice.start:data_slice.stop].std()
            stdev_y+=left_y_deg[data_slice.start:data_slice.stop].std()
            
            deltas_x=left_x_deg[data_slice.start+1:data_slice.stop]-left_x_deg[data_slice.start:data_slice.stop-1]
            deltas_y=left_y_deg[data_slice.start+1:data_slice.stop]-left_y_deg[data_slice.start:data_slice.stop-1]
            rms_x+=np.sqrt(np.sum(deltas_x*deltas_x)/len(deltas_x))
            rms_y+=np.sqrt(np.sum(deltas_y*deltas_y)/len(deltas_y))

    stdev_x=stdev_x/len(left_x_valid_data_periods)
    stdev_y=stdev_y/len(left_x_valid_data_periods)
    rms_x=rms_x/len(left_x_valid_data_periods)
    rms_y=rms_x/len(left_x_valid_data_periods)

    copy_left_pupil=left_pupil.copy()
    copy_left_pupil[~keep_left_all]=0
    left_velocity[missing_left_indexes[1:-1]]=np.NaN
    left_velocity_cleared=left_velocity.copy()
    left_velocity_cleared[~keep_left_all]=np.NaN
#    print 'noise: ',len(time[1:-1]),len(left_velocity),len(copy_left_pupil[1:-1])
    plotResults(trial_id,time[1:-1],(left_velocity,None,left_pupil[1:-1],left_velocity_cleared,None,copy_left_pupil[1:-1]),
                    None,
                    fig_title='Original Sample Velocity Trace. Bars Indicate Eye Data Loss.',
                    fig2_title='Pruned Periods of Data Collection where Velocity > Trial Median ( Bars are Pruned Areas, not Included in Noise Calculation ).',
                    ylabel='Velocity',ds1_label_pref='Degrees / Sec')
    
    fig = plt.figure()
    ax = fig.add_subplot(111)
    scatter(left_x_deg[keep_left_all],left_y_deg[keep_left_all],s=10, marker='o', c='r')
    ax.set_xlabel("Horizontal Angle")
    ax.set_ylabel("Verical Angle")
    plt.title("STDEV and RMS Based Eye Sample Noise Calculation (stdev: x={0},y={1}; RMS: x={2},y={3} ).".format(stdev_x,stdev_y,rms_x,rms_y)) 
    ax.grid(True)
    plt.show()
    
    
calcSampleStreamNoise()
```

#3. Eye Sample Filtering for Spatial Noise Reduction

## Purposes:
1. Reduction of high frequency noise from eye position data.
2. Decreasing velocity and accelleration noise in the eye signal to aid in improved Eye Event Detection (Saccades, Fixations, Blinks).

## Notes:
1. The requirements of the experimental paradigm being used, and the need for real-time eye data signals, will often influence which Filtering alogorithms can be used for a particular study.
2. There is usually a trade-off between the number of data points required by a filtering algorithm / the computational complexity in performing the filter and the quality of the filtered output.
3. An ideal filter would:
    1. require few sample points per calculation, or require only previously collected data points when filting the current point.
    2. be computationally inexpensive (i.e. fast to calculate)
    3. maintain real changes in eye signal positions, throughout the full biological oculomotor frequency range, while removing artificial signal fluctutions and artifacts.
4. I'm not aware of any filter that is /ideal/.

```python
# Some constants used across many of the example filters used here
#

#### USER PROVIDED CONSTANTS ####

# msec sampling interval of eye tracker
#
ET_SAMPLING_HZ=120.0

# The desired order of the filter to be used
# (when applicable)...
#
FILTER_ORDER=3

# Desired filtering window size in msec
# Specifying the filter size in msec allows
# for varying filter sample window size
# based on the ET sampling rate.
# > ET sampling rate == > filter sample window length
# for a given filter window duration.
# ( used when applicable )
#
FILT_WINDOW_MSEC=75.0

#### CALCULATED CONSTANTS ####

# Calculate filter window size in samples.
# ( used when applicable )
#
FILT_WIN_SAMPLE_CNT=int(FILT_WINDOW_MSEC//(1000.0/ET_SAMPLING_HZ))

# Ensure window sample size is odd.
#
if FILT_WIN_SAMPLE_CNT%2==0:
   FILT_WIN_SAMPLE_CNT+=1 

# Set sigma used in Gaussian Filter to == FILT_WIN_SAMPLE_CNT//3.5
#
SIGMA=FILT_WIN_SAMPLE_CNT//3.5

display('** Filter Sample Window Size: {0}'.format(FILT_WIN_SAMPLE_CNT))
```

#Some Common Filters

##Butterworth Filter

```python
# Cutoff Frequency, calculated based on the
# specified percentage of the 
# Nyquist frequency (ET_SAMPLING_HZ/2.0)
# 
NYQUIST_FREQ_PERC=50.0
TWO_PHASE=True

Wn=((ET_SAMPLING_HZ/2.0)*(NYQUIST_FREQ_PERC*0.01))/ET_SAMPLING_HZ
b, a = scipy.signal.butter(FILTER_ORDER, Wn, 'low')

if TWO_PHASE is True:
    ffunc=scipy.signal.filtfilt
else:
    ffunc=scipy.signal.lfilter
    
left_x_copy=left_x_deg.copy()
left_y_copy=left_y_deg.copy()
right_x_copy=right_x_deg.copy()
right_y_copy=right_y_deg.copy()

left_x_filtered = ffunc(b, a, left_x_deg)
left_y_filtered = ffunc(b, a, left_y_deg)
right_x_filtered = ffunc(b, a, right_x_deg)
right_y_filtered = ffunc(b, a, right_y_deg)

left_x_copy[missing_left_indexes]=np.NaN
left_y_copy[missing_left_indexes]=np.NaN
right_x_copy[missing_right_indexes]=np.NaN
right_y_copy[missing_right_indexes]=np.NaN
left_x_filtered[missing_left_indexes]=np.NaN
left_y_filtered[missing_left_indexes]=np.NaN
right_x_filtered[missing_right_indexes]=np.NaN
right_y_filtered[missing_right_indexes]=np.NaN    
    
    
plotResults(trial_id,time,(left_x_copy,left_y_copy,left_pupil,right_x_copy,right_y_copy,right_pupil),(left_x_filtered,left_y_filtered,right_x_filtered,right_y_filtered),
                         fig_title='Butterworth Filtered (Wn: {0}, order: {1}) vs Unfiltered Eye Sample Positions'.format(Wn,FILTER_ORDER),ylabel='Degrees of Visual Angle',ds1_label_pref='Unfiltered',ds2_label_pref='Butterworth Filtered')
```

#[Savitzky Golay Smoothing Filter](http://en.wikipedia.org/wiki/Savitzky%E2%80%93Golay_smoothing_filter)

_Quoting from wikipedia:_
    
The Savitzky–Golay smoothing filter is a filter that essentially performs a local polynomial regression (of degree k) on a series of values 
(of at least k+1 points which are treated as being equally spaced in the series) to determine the smoothed value for each point. 
The main advantage of this approach is that it tends to preserve features of the distribution such as relative maxima, minima and width, 
which are usually 'flattened' by other adjacent averaging techniques (like moving averages, for example).

This filtering algorithm was first described in 1964 by [Abraham Savitzky](http://en.wikipedia.org/wiki/Abraham_Savitzky) 
and [Marcel J. E. Golay](http://en.wikipedia.org/wiki/Marcel_J._E._Golay).

Savitzky and Golay's paper is one of the most widely cited papers in the [journal Analytical Chemistry](http://en.wikipedia.org/wiki/Analytical_Chemistry_(journal)) 
and is classed by that journal as one of its "10 seminal papers" saying "it can be argued that the dawn of the computer-controlled analytical instrument can be traced to this article"

### Python Implementation written by Dr. Thomas Haslwanter
**[SciPy / Numpy implementation from the SciPy Cookbook](http://www.scipy.org/Cookbook/SavitzkyGolay)**

#4a. Velocity Calculation

```python
def testVelocityCalc():
    lvx=calculateVelocity(time,left_x_deg)
    lvy=calculateVelocity(time,left_y_deg)
    rvx=calculateVelocity(time,right_x_deg)
    rvy=calculateVelocity(time,right_y_deg)

    lv=np.sqrt(lvx*lvx+lvy*lvy)
    rv=np.sqrt(rvx*rvx+rvy*rvy)

    weights=np.asarray([1.5,2.,3.,2.,1.5])
    weights=weights/np.sum(weights)
    
    lvx_filtered = calculateVelocity(time[2:-2],np.convolve(left_x_deg, weights,'valid'))
    lvy_filtered = calculateVelocity(time[2:-2],np.convolve(left_y_deg, weights,'valid'))
    rvx_filtered = calculateVelocity(time[2:-2],np.convolve(right_x_deg, weights,'valid'))
    rvy_filtered = calculateVelocity(time[2:-2],np.convolve(right_y_deg, weights,'valid'))
    lv_filtered=np.sqrt(lvx_filtered*lvx_filtered+lvy_filtered*lvy_filtered)
    rv_filtered=np.sqrt(rvx_filtered*rvx_filtered+rvy_filtered*rvy_filtered)

    lv_copy=lv.copy()
    rv_copy=rv.copy()
    
    lv_copy[missing_left_indexes[1:-1]]=np.NaN
    lv_filtered[missing_left_indexes[3:-3]]=np.NaN
    rv_copy[missing_right_indexes[1:-1]]=np.NaN
    rv_filtered[missing_right_indexes[3:-3]]=np.NaN
 

    plotResults(trial_id,time[3:-3],(lv_copy[2:-2],None,left_pupil[3:-3],rv_copy[2:-2],None,right_pupil[3:-3]),(lv_filtered,None,rv_filtered,None),fig_title='Velocity (degrees / sec) Data',ylabel='Velocity',ds1_label_pref='Unfiltered Velocity Data',ds2_label_pref='+/- 2 Sample Weighted Velocity Filter')

testVelocityCalc()
```

#4b. Accelleration Calculation

```python
def testAccellerationCalc():
    lax=calculateAccelleration(time,left_x_deg)
    lay=calculateAccelleration(time,left_y_deg)
    rax=calculateAccelleration(time,right_x_deg)
    ray=calculateAccelleration(time,right_y_deg)

    la=np.sqrt(lax*lax+lay*lay)
    ra=np.sqrt(rax*rax+ray*ray)
    la_copy=la.copy()
    ra_copy=ra.copy()

###################

    weights=np.asarray([1.5,2.,3.,2.,1.5])
    weights=weights/np.sum(weights)
    
    lax_filtered = calculateAccelleration(time[2:-2],np.convolve(left_x_deg, weights,'valid'))
    lay_filtered = calculateAccelleration(time[2:-2],np.convolve(left_y_deg, weights,'valid'))
    rax_filtered = calculateAccelleration(time[2:-2],np.convolve(right_x_deg, weights,'valid'))
    ray_filtered = calculateAccelleration(time[2:-2],np.convolve(right_y_deg, weights,'valid'))

    la_filtered=np.sqrt(lax_filtered*lax_filtered+lay_filtered*lay_filtered)
    ra_filtered=np.sqrt(rax_filtered*rax_filtered+ray_filtered*ray_filtered)


    la_copy[missing_left_indexes[2:-2]]=np.NaN
    la_filtered[missing_left_indexes[4:-4]]=np.NaN
    ra_copy[missing_right_indexes[2:-2]]=np.NaN
    ra_filtered[missing_right_indexes[4:-4]]=np.NaN
 

    plotResults(trial_id,time[4:-4],(la_copy[2:-2],None,left_pupil[4:-4],ra_copy[2:-2],None,right_pupil[4:-4]),(la_filtered,None,ra_filtered,None),fig_title='Accelleration (degrees / sec**2) Data',ylabel='Accelleration',ds1_label_pref='Unfiltered Accelleration Data',ds2_label_pref='+/- 2 Sample Weighted Accelleration Filter')


testAccellerationCalc()
```
